```python
df = pd.read_csv('SalaryData.csv')
df
```

```python
#4a. Какие атрибуты высококоррелированы, определить характер корреляции
#Берем значение корреляции по модулю:
#до 0,2: Очень слабая корреляция
#до 0,5: Слабая корреляция
#до 0,7: Средняя корреляция
#до 0,9: Высокая корреляция
#свыше 0,9: Очень высокая корреляция

#4b. Какие атрибуты не имеют корреляцию
# Корреляция отсутствует у тех атрибутов, у которых коэфф. корр. равен или примерно равен нулю 0. Нулевая корреляция говорит 
#о том что данный регрессор не влияет на зависимую
```

```python
#4с.Определить корреляцию между параметрами
#Для рассчета корреляции уберем атрибуты: Gender, Education level, Job Title(так как их тип - строчный)

df_new = df[['Age','Years of Experience','Salary']]
corr_matrix = df_new.corr()
sns.heatmap(corr_matrix, annot=True, cmap='cool')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df_new)
plt.title('Pairplot of Numerical Variables')
plt.show()
```

```python
#4d. Выводы: В этой работе проведено исследование коэфф-ов корреляции, 
#а также соответствующие графики.
#
#Какие выводы о коэффиценте корреляции можно сделать?
#Чем точнее один атрибут переходит в другой(речь о линейной зависимости), тем выше 
#коэффицент корреляции.
#
#Положетельный коэфф. корреляции означает, что зависимость атрибутов возрастает(
#возрастает один, возрастает и другой)
#
#Отрицательный коэфф. корреляции означает, что зависимость атрибутов убывает(
#один возрастает, другой убывает)
```
